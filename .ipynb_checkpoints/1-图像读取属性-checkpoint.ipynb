# 图像读取

## using skimage
https://cloud.tencent.com/developer/section/1414650

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import io
from skimage import data
```

```python
# im = io.imread('tst.png', as_gray=False)
im = data.astronaut()
io.imshow(im)
```

## using opencv

```python
import cv2
```

```python
im = cv2.imread('data/lena.png')
# display directly
# io.imshow(im)

# bgr->rgb
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
io.imshow(im)
```

## 图像属性

### 图像是矩阵-skimage

```python
from skimage import color
```

```python
img = color.rgb2gray(im)
io.imshow(img)
print(img[:10,:10])
```

```python
img =cv2.cvtColor(im, cv2.COLOR_RGB2GRAY)
# io.imshow(img)
print(img[:10,:10])
```

### 图像长，宽，通道（位面）

```python
h,w,n= im.shape
print(f'image height:{h}, width:{w}, channels:{n}')
```
