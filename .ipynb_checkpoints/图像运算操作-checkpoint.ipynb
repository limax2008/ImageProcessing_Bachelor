## 空间几何操作

### 位移

M=[[1 0 t_x];[0 1 t_u]]

```python
img = cv2.cvtColor(im, cv2.COLOR_RGB2GRAY)
cols,rows = img.shape
# 定义位移矩阵
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(rows, cols))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')

plt.subplot(1,2,2)
plt.imshow(dst,cmap='gray')
```

### 旋转

 Rotation transformation
The transformation matrix:
 M=$
 \left|\begin{matrix}
   cos(\theta) & sin(\theta) & 0 \\
    -sin(\theta) & cos(\theta) & 0 \\
    0 & 0 & 1
   \end{matrix} \right|
 $

In opencv, the matrix M can be written as :
M=
$
 \left|\begin{matrix}
  \alpha & \beta & (1-\alpha) \cdot center.x-\beta \cdot center.y \\
    -\beta & \alpha & \beta \cdot center.x+(1-alpha) \cdot center.y \\
    %0 & 0 & 1
   \end{matrix} \right|
 $
 
where $\alpha =scale \cdot cos(\theta)$ and $\beta= scale \cdot sin(\theta)$ are the scaling factor. 

<!-- The center is usually defined as the image center.  -->
**OPENCV uses cv2.getRotationMatrix2D to obtain the transformation matrix.**

```python
h,w = img.shape
M = cv2.getRotationMatrix2D((h/2.0,w/2.0),60, 0.8)
dst = cv2.warpAffine(img, M,(w,h))

plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')

plt.subplot(1,2,2)
plt.imshow(dst,cmap='gray')
```

### 尺度变换

 M=$
 \left|\begin{matrix}
    d_x & 0 & 0 \\
    0 & d_y & 0 \\
    0 & 0 & 1
   \end{matrix} \right|
 $
 
 M  =$\left|\begin{matrix}
    d_x & 0 & 0 \\
    0 & d_y & 0 
   \end{matrix} \right|
 $

```python
# 定义尺度变换矩阵
dx=1.8
dy =1
M = np.float32([[dx,0,0],[0,dy,0]])
dst = cv2.warpAffine(img, M, (int(w*dx),h))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')

plt.subplot(1,2,2)
plt.imshow(dst,cmap='gray')
```

### 偏移

 M=$
 \left|\begin{matrix}
    1 & 0 & 0 \\
    s_v & 1 & 0 \\
    0 & 0 & 1
   \end{matrix} \right|
 $
 
 M  =$\left|\begin{matrix}
    1 & 0 & 0 \\
    s_v & 1 & 0 
   \end{matrix} \right|
 $

```python
# 定义垂直偏移矩阵
sv=1.1
M = np.float32([[1,0,0],[sv,1,0]])
dst = cv2.warpAffine(img, M, (w,h*2))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')

plt.subplot(1,2,2)
plt.imshow(dst,cmap='gray')
```

```python
# 定义水平偏移矩阵
sv=1.1
M = np.float32([[1,sv,0],[0,1,0]])
dst = cv2.warpAffine(img, M, (w*2,h*2))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')

plt.subplot(1,2,2)
plt.imshow(dst,cmap='gray')
```

```python
# 定义水平+垂直偏移矩阵
sv=0.5
M = np.float32([[1,sv,0],[sv,1,0]])
dst = cv2.warpAffine(img, M, (w*2,h*2))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')

plt.subplot(1,2,2)
plt.imshow(dst,cmap='gray')
```
