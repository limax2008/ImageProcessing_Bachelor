```python
import matplotlib.pyplot as plt
```

```python
x= np.random.random(10)
y = np.arange(10)
plt.plot(y,x,color ='r', linestyle =':')
```

## scatter 散点图

```python
x= np.arange(10)
y = np.random.random(10)
plt.scatter(x,y, color='r', linewidths=10, marker='*')
```
